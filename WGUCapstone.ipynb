### iv. Cleaning, Parsing, and Featurizing Dataset
---
Luckily, the mushrooms.csv dataset is already cleaned in the sense that it contains no erroneous or null values.

Featurizing the dataset is a broad practice, involving data transformation and engineering to draw more utility out of data. The great number of features included in the dataset could mean that engineering new features isn't necessary for this model to perform. I will be performing featurization in the process of converting all the data to ordinal data and removing extraneous features, however, shown below.

Parsing of the data also takes place below in which the class will be separated from input features before a test/train split is performed.

````python
labelencoder = LabelEncoder()
for column in df:
  df[column] = labelencoder.fit_transform(df[column])
df.head()
````

When the above cell is run, the altered numerical data is generated and the first 5 members of the dataset are displayed.

````python
x = df.drop('class', axis=1)
y = df['class']
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42, test_size=0.5)
RF = RandomForestClassifier()
RF.fit(x_train, y_train)
importances = RF.feature_importances_
sorted_importances = np.argsort(importances)
features = x.columns.values
plt.figure(figsize=(10, 10))
plt.barh(range(len(features)), importances[sorted_importances], align='center' )
plt.yticks(range(len(features)), features[sorted_importances])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances')

````

Above, the data is randomly split into 2 sets: one for training, and one for testing. The test data will remain isolated from the model until testing to ensure that the results of the final test are as fair as possible. A Random Forest Classifier model is trained and used alongside the feature_importances function to evaluate the significance of the correlation between each trait and inedibility. Using this information, we can test selections of a handful of traits and evaluate which configuration is ideal.

As demonstrated above, Odor, Gill Color, Gill Size, Spore Print Color, Bruises, Population, and Ring Type are all attributes that hold a relatively high weight in determining the class of a given mushroom. Using some combination of these attributes, potentially including an attribute or two with lower significance, should give us the classification accuracy desired.

````python
test_set1 = x[["odor", "gill-color", "gill-size", "bruises", "population"]].copy()
test_set2 = x[["gill-color", "gill-size", "stalk-root", "odor", "population"]].copy()
test_set3 = x[["odor", "gill-color", "habitat", "stalk-root", "population"]].copy()
test_set4 = x[["odor", "bruises", "stalk-root", "population"]].copy()
train_x1, test_x1, train_y1, test_y1 = train_test_split(test_set1, y, random_state = 42, test_size=0.5)
train_x2, test_x2, train_y2, test_y2 = train_test_split(test_set2, y, random_state = 42, test_size=0.5)
train_x3, test_x3, train_y3, test_y3 = train_test_split(test_set3, y, random_state = 42, test_size=0.5)
train_x4, test_x4, train_y4, test_y4 = train_test_split(test_set4, y, random_state = 42, test_size=0.5)
test_rf1 = RandomForestClassifier()
test_rf2 = RandomForestClassifier()
test_rf3 = RandomForestClassifier()
test_rf4 = RandomForestClassifier()
test_rf1.fit(train_x1, train_y1)
test_rf2.fit(train_x2, train_y2)
test_rf3.fit(train_x3, train_y3)
test_rf4.fit(train_x4, train_y4)
test_preds1 = test_rf1.predict(test_x1)
test_preds2 = test_rf2.predict(test_x2)
test_preds3 = test_rf3.predict(test_x3)
test_preds4 = test_rf4.predict(test_x4)
test_matrix1 = confusion_matrix(test_y1, test_preds1)
test_matrix2 = confusion_matrix(test_y2, test_preds2)
test_matrix3 = confusion_matrix(test_y3, test_preds3)
test_matrix4 = confusion_matrix(test_y4, test_preds4)
print(test_matrix1)
print(test_matrix2)
print(test_matrix3)
print(test_matrix4)
````

Above, 4 test models are trained on different subsets of the data then tested on the same set.
